```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.33, random_state=10)
```

Feature Engineering
Standardize the dataset or feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scalar
```

IMPORTANT: USE fit_tranform() FOR TRAIN DATA AND USE transform() on TEST data

When we do fit_transform():
1.it calcualtes the mean and std dev from the train data 
2.scales down the feature 
3.applies it to the test data.
We do not use fit_transform() on the test data because it will again calculate the mean and the std dev from the test data and will lead to data leakage.


```python
X_train=scalar.fit_transform(X_train)
```

```python
X_test=scalar.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train,y_train)
```

```python
##print the coefficient and intercept
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

PREDICTION FOR THE TEST DATA

```python
reg_pred=regression.predict(X_test)
```

ASSUMPTIONS OF LINEAR REGRESSION

```python
plt.scatter(y_test,reg_pred)
plt.xlabel("test truth data")
plt.ylabel("test predicted data")
```

RESIDUAL=to get errors

```python
residuals=y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

Scatter plot with predictions and residuals
Uniform distribution

```python
plt.scatter(reg_pred,residuals)
```

Performance Metrics

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

R SQUARE AND ADJUSTED R SQUARE

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

ADJUSTED R SQUARE

```python
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
